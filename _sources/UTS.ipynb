### Text Prepocessing

```python
#import library
import re
import string
```

```python
#text cleaning
def text_cleaning(text):
  #hapus url
  text = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});').sub('', str(text))
 
  #ubah huruf kecil
  text = text.lower()

  #hapus karakter awal dan akhir
  text = re.strip()

  #hapus tanda hubung(punctuation)
  text = re.compile('<.*?>').sub("",text)

  #hapus karakter spesial
  text =re.compile("[%s]" % re.escape(string.punctuation))
  
  #hapus spasi ganda
  text = re.sub("\s+","",text)

  #hapus angka
  text = re.sub(r"\[[0-9]*\]","",text)
  text = re.sub(r"[^\w\s]","",str(text).lower().strip())
  text = re.sub(r"\d","",text)
  text = re.sub(r"\s","",text)

  #ubah text nan menjadi whitespace
  text = re.sub("nan","",text)

  return text
```

```python
import re

def text_cleaning(text):
  #hapus karakter yang tidak diperlukan
  text = re.sub(r'[^\w\s]','',text)
  
  #hapus angka
  text = re.sub(r'\d+','',text)
  
  #ubah ke huruf kecil
  text = text.lower()
  
  #hapus karakter awal dan akhir
  text = text.strip()
  
  #hapus tanda hubung(punctuation)
  text = re.sub(r'[\-]',' ',text)
  
  #hapus multiple spasi
  text = re.sub(r'\s+',' ',text)
  
  return text
```

### Tokenization

### Stopwords Removal

```python
#menghapus kata yang tidak digunakan
from nltk.corpus import stopwords
from itertools import chain
nltk.download("stopwords")
```

```python
stopword = set(chain(stopwords.words("indonesian"), stopwords.words("english")))

data["judul_tokens"]=data["judul_tokens"].apply(lambda x:[w for w in x if not w in stopword])
data["abstrak_tokens"]=data["abstrak_tokens"].apply(lambda x:[w for w in x if not w in stopword])
```

```python
data[["Judul","judul_tokens"]].head()
```

```python
data[["Abstrak", "abstrak_tokens"]].head()
```

### Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from tqdm.auto import tqdm
tqdm.pandas()

factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Stemming judul
data['judul_tokens'] = data['judul_tokens'].progress_apply(lambda x: stemmer.stem(' '.join(x)).split(' '))
```

```python
#hasil stemming judul
data[["Judul","judul_tokens"]].head()
```

```python
# Stemming abstrak
data['abstrak_tokens'] = data['abstrak_tokens'].progress_apply(lambda x: stemmer.stem(' '.join(x)).split(' '))
```

```python
#hasil stemming abstrak
data[["Abstrak","abstrak_tokens"]].head()
```
