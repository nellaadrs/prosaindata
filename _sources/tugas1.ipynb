## Data Twitter

```python
import snscrape.modules .twitter as sntwitter
import pandas as pd

#array tweet
list_tweet= []

#mengambil data rentang waktu tertentu 
for data, tweet in enumerate(sntwitter.TwitterSearchScraper("prabowo since:2022-01-01 until:2022-02-22").get_items()):
  if data >=500:
    break
  list_tweet.append([tweet.user.username,tweet.content,tweet.user.location])
```

```python
data_twitter =pd.DataFrame(list_tweet, columns=["Username","Konten","Lokasi"])
data_twitter
```

```python
data_twitter.to_csv("data_TA.csv")
```

## Mencari data PTA

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd 
import csv

#permintaan dari web browser ke web server untuk menampilkan sesuai dengan SO yang digunakan
browser = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.5481.104 Safari/537.36"}
```

```python
def scrape_TA(page):
  global browser
  data=[]
  for laman in range(1,page+1):
    if laman==1 :
      URL = "https://pta.trunojoyo.ac.id/c_search/byprod/10/"
    else :
      URL = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{laman}"
    request = req.get(URL,browser).text
    #var bs yang menyimpan data request berupa html
    soup = bs(request, 'lxml')
    prodi = soup.find_all('div',{'id':'begin'})
    for pro in prodi:
      prod = pro.find('h2').text
    jurusan = prod[-18:]
    ul = soup.find('ul', 'items list_style')
    li = ul.find_all('li', {'data-id':'id-1'})
    for x in li:
      link = x.find('a','gray button')['href']
      request2 = req.get(link, browser).text
      soup2 = bs(request2, 'lxml')
      abstrak= soup2.find('p',{'align':'justify'}).text.replace('\r','').replace('\n','')
      NPM = x.find('a','gray button')['href'][-12:]
      headline = x.find('a', 'title').text.replace('\r','').replace('\n','')
      data.append([NPM,headline,abstrak,jurusan])
  return data
```

```python
data_TA= scrape_TA(100)
data_TA_result=pd.DataFrame(data_TA)
data_TA_result.columns = ["NPM", "Judul","Abstrak","jurusan"]
data_TA_result
```

```python
data_TA_result.to_csv("data_TA.csv")
```

## Mencari Data Berita online

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd 
import csv

#permintaan dari web browser ke web server untuk menampilkan sesuai dengan SO yang digunakan
browser = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.5481.104 Safari/537.36"}
```

```python
def scrape_detik (laman):
  global browser
  data=[]
  for page in range (1, laman+1):
    #print (page) 
    # mengambil data banjir kemudian diurutkan berdasarkan waktu dan halaman
    url = f"https://www.detik.com/search/searchall?query=banjir&siteid=2&sortby=time&page={laman}"
    req_get = req.get(url,browser).text
    # memanggil beautysoup
    bsoup = bs(req_get, "lxml")
    #menampilkan list berita
    list_berita = bsoup.find("div","list media_rows list-berita")
    # menampil isi artikel
    artikel_berita = list_berita.find_all("article") 

    for tampil in artikel_berita:
      #judul
      link = tampil.find_all("h2")
      link = [ele.text.strip() for ele in link]
      # deskripsi
      isi  = tampil.find_all("p")
      isi  = [ele.text.strip() for ele in isi]
      data.append([link[0], isi[0]])
  return data
```

```python
berita=scrape_detik(20)
hasil_berita = pd.DataFrame(berita)
hasil_berita.columns=["Judul","Isi KOnten"]
hasil_berita
```

```python
hasil_berita.to_csv("berita_banjir.csv")
```
